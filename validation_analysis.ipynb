```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as nps
import os
```

```python
rootdir = 'D:/CODING/erp_contest/hull_tactical'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if '-results.csv' in file:
            results = pd.read_csv(os.path.join(subdir, file))
            plot = plt.figure()
            info = plot.add_subplot(111)
            plot.subplots_adjust(top=0.85)
            actual, = plt.plot(results['actual'])
            prediction, = plt.plot(results['prediction'])
            plt.title(file)
            plt.legend([actual, prediction], ["actual", "prediction"])  
            plt.figure()
            plt.title(file +' loss')
            plt.plot(results["loss"])
            ave_loss = results["loss"].mean()
            info.text(len(results['loss']) + 25 ,3, 'Average loss: ' + str(ave_loss))
            R_bar = results["actual"].mean()
            bot = (results["actual"]-R_bar)**2
            bot = bot.sum()
            r_hat = results["prediction"]
            top = (results["actual"]-r_hat)**2
            top = top.sum()
            r_2 = top/bot
            info.text(len(results['loss'])+25, 2, 'R squared: ' + str(r_2))
    
            
```
